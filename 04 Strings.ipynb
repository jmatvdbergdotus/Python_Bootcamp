strings are ordered sequences, it means that we can use indexing and slicing to grab sub-sections of the string

```python
x = "hello"
print(x[2])
```

indexing uses [ ] notation after the string  and allows you to grab a single character from the string. starts from 0. can also be reverse indexing 0, -1

slicing allows to grab a sub-section of multiple characters. this has the ff syntax [start:stop:step]
start - numerical index for the slice start
stop - the index you will go up to (but not include)
step - is the size of the jump  you take

String formatting for printing. .format() method and the f-strings (formatted strings literals)

```python
# .format() method
```

```python
print("I am {}".format("BEAUTIFUL"))
```

```python
print("I am {} {} {}".format("OUT","INSIDE","BEAUTIFUL"))
```

```python
print("I am {2} {1} {0}".format("OUT","INSIDE","BEAUTIFUL"))
```

```python
print("I am {b} {i} {o}".format(o="OUT",i="INSIDE",b="BEAUTIFUL"))
```

```python
# FLoat formatting follows "{value:width.precision f}"
```

```python
result = 100/777
```

```python
result
```

```python
print('The result was {r:1.3f}'.format(r=result))
```
