```python
hungry = True
if hungry:
    print ("FEED ME!")
else:
    print ("I'm not hungry!")
```

```python
loc = 'Bank'
if loc == 'Auto Shop':
    print ("Cars are cool!")
else:
    print ("I don't know much")
```

```python
loc = 'Bank'
if loc == 'Auto Shop':
    print ("Cars are cool!")
elif loc == 'Bank':
    print ("Money is cool!")
elif loc == 'Store':
    print ("Welcome to my store!")
else:
    print ("I don't know much")
```
